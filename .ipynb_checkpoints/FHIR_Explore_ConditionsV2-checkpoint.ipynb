# PcorNet

```python
pcornet = pd.read_csv('input/pcornet/diagnosis_fhir.csv')
```

```python
pcornet
```

```python
fhir = fhirUtil.conditions[fhirUtil.conditions['coding_system'] == 'ICD10']
```

```python
m = fhir.drop_duplicates(subset="user_id").merge(pcornet.drop_duplicates(subset="user_id"), on="user_id", how="outer", indicator=True)
m['_merge'].value_counts(dropna=False)
```

```python
def compare(left, right, config, search_icd10_codes, left_field, right_field, verbose=True):
#         {
#         'title': 'Diabetes', 
#         'codes': ['E10','E11'], 
#         'ccs_field': 'diabetes', 
#         'ccs_yes': [1], 
#         'ccs_no': [2]
#     },
    left[left[left_field].isin()]
    r = {
        'both_len': 0,
        'left_only_len': 0,
        'right_only_len': 0,
        'both_negative_len': 0,
        'total_yes': 1,
        'total_no': 1
    }
    return r
    
    
result = []
for config in comparison_config:
    print('Processing ', config['title'])
    comp = compare(
                               left=fhir,
                               right=pcornet,
                               config=config,
                               search_icd10_codes=config['codes'], 
                               left_field= 'coding_code',
                               right_field='dx',
                               verbose=False)

    resultRow = {
    'title': config['title'],
    'Both Pos': comp['both_len'],
    'CF %': round((comp['both_len']*100)/comp['total_yes']),
    'FHIR Pos': comp['left_only_len'],
    'C %': round((comp['left_only_len']*100)/comp['total_yes']),
    'PCORNET Pos': comp['right_only_len'],
    'F %': round((comp['right_only_len']*100)/comp['total_yes']),
    'Both Negative': comp['both_negative_len'],
    'BN %': round((comp['both_negative_len']*100)/(comp['total_no']))
    }
    result.append(resultRow)
result_df = pd.DataFrame(result)
result_df
```
